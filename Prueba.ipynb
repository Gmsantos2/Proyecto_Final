```python
import pandas as pd #importaciones
import numpy as np
from bokeh.plotting import figure, output_file, output_notebook, show
from bokeh.palettes import *
from math import pi
from bokeh.transform import cumsum 
output_notebook() #output que presenta los cambios del bokeh
datos = pd.read_csv("data/lugares.csv",sep=",") # variable donde leemos nuestro archivo csv
datos #presentamos los datos
```

```python
group = datos.groupby('Nombre', as_index = False).mean() 
x_range= group['Nombre']
y_range= group['Latitud']
colores = viridis(len(datos))
grafico = figure(x_range=x_range,plot_height=600, plot_width = 2000, title="Latidud por Lugar turisctico", toolbar_location=None, tools="")
grafico.vbar(x=x_range, top=group['Latitud'], width=0.9, line_width=3,line_color = 'white',fill_color = colores)
grafico.y_range.start = -29.5
grafico.yaxis.axis_label = "NumeroLugares"
grafico.xaxis.axis_label = "Lugar"
grafico.xaxis.major_label_orientation = 1.55
show(grafico)
```

```python
group1 = datos.groupby('Nombre', as_index = False).mean() 
subdatos=datos["Longuitud"]
x_range= group1['Nombre']
y_range= subdatos
colores = inferno(len(datos))
grafico2 = figure(x_range=x_range,plot_height=600, plot_width = 2000, title="Longitud por Lugar turisctico", toolbar_location=None, tools="")
grafico2.vbar(x=x_range, top=subdatos, width=0.9, line_width=3,line_color = 'white',fill_color = colores)
grafico2.y_range.start = -100
grafico2.yaxis.axis_label = "Longitud"
grafico2.xaxis.axis_label = "Lugar"
grafico2.xaxis.major_label_orientation = 1.55
show(grafico2)
```

```python
x = datos.groupby(["Provincia"])["Nombre"].count()
data = pd.Series(x).reset_index(name='value').rename(columns={'index':'Nombre'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = inferno(len(x))
p = figure(plot_height=600,plot_width = 900, title="Numero de lugares.T por provincia",
           tools="hover", tooltips="@Provincia: @value", x_range=(-0.5, 1.5))
p.wedge(x=0, y=1, radius=0.5,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="black", fill_color='color', legend="Provincia", source=data)
p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None
show(p)


```

```python
x = datos.groupby(['TipoLugar'])['TipoLugar'].count()
data = pd.Series(x).reset_index(name='value').rename(columns={'index':'TipoLugar'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = inferno(len(x))
p = figure(plot_height=600,plot_width = 900, title="Grafico de tipo de lugares por  lugar turisticos",
           tools="hover", tooltips="@TipoLugar: @value", x_range=(-0.5, 1.5))
p.wedge(x=0, y=1, radius=0.5,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="black", fill_color='color', legend="TipoLugar", source=data)
p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None


show(p)
```
